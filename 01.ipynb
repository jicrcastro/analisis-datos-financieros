```python
#Importación de librerías
import numpy as np
import pandas as pd
```

```python
#Lectura de archivo SPSS y de columnas especificas
df = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\Colaboradores_202206.sav', usecols=['CTARUT'])
#Lo convierto a dataframe, porque viene como series al ser una columna
df = pd.DataFrame(df)
df
```

```python
# cambio de nombre de columna de CTARUT a RUT
df = df.rename(columns={'CTARUT':'RUT'})
df
```

```python
#Cambio tipo de dato de columna a entero
df['RUT'] = df['RUT'].astype(int)
df
```

```python
#Lectura archivo csv
df2 = pd.read_csv('C:\\empresa\\analisis 02\\actualizadosv3\\CLASIFDI20220725.csv', usecols=['CTARUT'], sep=';')
df2
```

```python
#Cambio de nombre de columna de CTARUT a RUT
df2 = df2.rename(columns={'CTARUT':'RUT'})
df2
```

```python
#Creacion de columna cliente con valor 1
df2['Cliente'] = 1
df2
```

```python
#Left join de ambos dataframes, para que me queden solo los rut que estan en el archivo SPSS con los ruts que coincidentes del archivo csv, agregando la columna cliente
df3 = pd.merge(df, df2, on='RUT', how='left')
df3.head(10)
```

```python
#Reemplazo los valores NaN por 0 y cambio el tipo de dato de la columna cliente a entero
df3['Cliente'] = df3['Cliente'].fillna(0)
df3['Cliente'] = df3['Cliente'].astype(int)
df3
```

```python
#Lectura de archivo SPSS y de columnas especificas
df4 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\Oferta_Abierta_Preevaluados_202208.sav', usecols=['RUT', 'CUPO'])
df4.head(10)
```

```python
# creo una columna de nombre Tipo_Oferta con valor "Otro"
df4['Tipo_Oferta'] = "Otro"
df4
```

```python
# Hago un merge con los dataframes df3 y df4
df5 = pd.merge(df3, df4, on='RUT', how='left')
df5.head(10)
```

```python
# cambio el tipo de dato de la columna CUPO a entero
df5['CUPO'] = df5['CUPO'].astype(int)
df5.head(10)
```

```python
#Lectura de archivo SPSS y cambio de nombre de columna CTARUT a RUT
df6 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\BASES_NO_CALIFICA_20220714.sav')
df6 = df6.rename(columns={'CTARUT':'RUT'})
df6.head(10)
```

```python
#Hago un merge de los df5 y df6 para que me agreguen mas informacion a los ruts coincidentes, en este caso la columna razón
df7 = pd.merge(df5, df6, on='RUT', how='left')
df7.head(10)
```

```python
df8 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\RESUMEN_5 años_20220725.sav', usecols=['CTARUT', 'MTO_TOT_BIC', 'MTO_TOT_MCC', 'MTO_TOT_BIC_EQFX', 'MTO_TOT_MCC_EQFX'])
df8 = df8.rename(columns={'CTARUT':'RUT'})
df8
```

```python
#Hago un merge de los df7 y df8 para que me agreguen mas informacion a los ruts coincidentes, en este caso las columnas de monto
df9 = pd.merge(df7, df8, on='RUT', how='left')
df9.head(10)
```

```python
#Reemplazo los valores NaN por 0 en las 4 columnas de monto
df9[['MTO_TOT_BIC', 'MTO_TOT_MCC', 'MTO_TOT_BIC_EQFX', 'MTO_TOT_MCC_EQFX']] = df9[['MTO_TOT_BIC', 'MTO_TOT_MCC', 'MTO_TOT_BIC_EQFX', 'MTO_TOT_MCC_EQFX']].fillna(0)
df9
```

```python
# Creo una funcion para que me agregue una columna con el valor "Si" si el monto de alguna de las 4 columnas de deuda equifax es mayor a 0, y "No" si no lo es
#Luego aplico la funcion a cada fila del dataframe
def mora_eqfx_hoy(monto):
    if monto['MTO_TOT_BIC'] > 0 or monto['MTO_TOT_MCC'] > 0 or monto['MTO_TOT_BIC_EQFX'] > 0 or monto['MTO_TOT_MCC_EQFX'] > 0:
        return 'Si'
    else:
        return 'No'

df9['Mora_EQFX_Hoy'] = df9.apply(mora_eqfx_hoy, axis=1)
df9.head(10)
```

```python
# Lectura de archivo SPSS y de columnas especificas
df10 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\R04 Sin LMC 202205.sav', usecols=['RUT', 'Deuda_DMora', 'Deuda_DVencida', 'Deuda_DCastigo'])
df10.head(10)
```

```python
# Creo una funcion para que me agregue una columna con el valor "Si" si el monto de alguna de las 3 columnas de deuda de R04 es mayor a 0, y "No" si no lo es
#Luego aplico la funcion a cada fila del dataframe
def deuda_mora(deuda):
    if deuda['Deuda_DMora'] > 0 or deuda['Deuda_DVencida'] > 0 or deuda['Deuda_DCastigo'] > 0:
        return 'Si'
    else:
        return 'No'
    
df11['Mora_R04_Hoy'] = df11.apply(deuda_mora, axis=1)
df11.head(10)
```
