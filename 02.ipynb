```python
#lectura de archivo SPSS y de columnas especificas y luego cambio de nombre de columna CTARUT a RUT y comvierto el tipo de dato a int
df22 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\GSE_FAM_2022_02.sav', usecols=['CTARUT', 'NVO_GSE', 'GSE_FAM'])
df22 = df22.rename(columns={'CTARUT':'RUT'})
df22['RUT'] = df22['RUT'].astype(int)
df22.head(10)
```

```python
#Hago un merge de los df21 y df22 para que me agreguen mas informacion a los ruts coincidentes, en este caso las columnas
df23 = pd.merge(df21, df22, on='RUT', how='left')
df23.head(10)
```

```python
#lectura de archivo SPSS y de columnas especificas y luego cambio de nombre de columna CTARUT a RUT, MORA_EQ_sum_12_Meses a Mora_EQ_12_Meses y comvierto el tipo de datos a int
df24 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\mora_equifax_12_meses.sav')
df24 = df24.rename(columns={'CTARUT':'RUT', 'MORA_EQ_sum_12_Meses':'Mora_EQ_12_Meses'})
df24[['RUT', 'Mora_EQ_12_Meses']] = df24[['RUT', 'Mora_EQ_12_Meses']].astype(int)
df24.head(10)
```

```python
#Hago un merge de los df23 y df24 para que me agreguen mas informacion a los ruts coincidentes, en este caso las columnas
#Luego reemplazo los valores NaN por 0 en la columna Mora_EQ_12_Meses y cambio el tipo de dato a int
df25 = pd.merge(df23, df24, on='RUT', how='left')
df25['Mora_EQ_12_Meses'] = df25['Mora_EQ_12_Meses'].fillna(0)
df25['Mora_EQ_12_Meses'] = df25['Mora_EQ_12_Meses'].astype(int)
df25.head(10)
```

```python
# lectura de archivo SPSS y de columnas especificas y luego cambio de nombre de columna 
# CTARUT a RUT, Deuda_DMora_SUM a Deuda_DMora_12M, Deuda_DVencida_SUM a Deuda_DVencida_12M, Deuda_DCastigo_SUM a Deuda_DCastigo_12M
df26 = pd.read_spss(path='C:\\empresa\\analisis 02\\actualizadosv3\\mora_r04_12_meses.sav')
df26 = df26.rename(columns={'CTARUT':'RUT', 'Deuda_DMora_SUM':'Deuda_DMora_12M', 'Deuda_DVencida_SUM':'Deuda_DVencida_12M', 'Deuda_DCastigo_SUM':'Deuda_DCastigo_12M'})
df26.head(10)
```

```python
#Hago un merge de los df25 y df26 para que me agreguen mas informacion a los ruts coincidentes, en este caso las columnas.
#Luego reemplazo los valores NaN por 0 en varias columnas y cambio el tipo de dato a int
df27 = pd.merge(df25, df26, on='RUT', how='left')
df27[['Deuda_DMora_12M', 'Deuda_DVencida_12M', 'Deuda_DCastigo_12M']] = df27[['Deuda_DMora_12M', 'Deuda_DVencida_12M', 'Deuda_DCastigo_12M']].fillna(0)
df27[['Deuda_DMora_12M', 'Deuda_DVencida_12M', 'Deuda_DCastigo_12M']] = df27[['Deuda_DMora_12M', 'Deuda_DVencida_12M', 'Deuda_DCastigo_12M']].astype(int)
df27.head(10)
```

```python
df27['fecha_analisis'] = dt.datetime(2022,7,1)
df27['Dif_Fechas'] = round(((df27.fecha_analisis - df27.Periodo_Min) / np.timedelta64(1, 'M')).fillna(0))
df27['Dif_Fechas'] = df27['Dif_Fechas'].astype(int)
df27.head(10)
```

```python
#tramifico la columna Dif_Fechas en tramos de castigos por meses, luego la aplico a todo el df27 retornando los resultado en la columna Tramos_Castigos
def tramos_castigos(fechas):
    if fechas['Dif_Fechas'] > 0 and fechas['Dif_Fechas'] <= 6:
        return '<= 6 meses'
    elif fechas['Dif_Fechas'] > 6 and fechas['Dif_Fechas'] <= 12:
        return 'Entre 6 a 12 Meses'
    elif fechas['Dif_Fechas'] > 12 and fechas['Dif_Fechas'] <= 24:
        return 'Entre 1 a 2 Años'
    elif fechas['Dif_Fechas'] > 24:
        return 'Mayor a 2 Años'
    else:
        return 'No Castigado'

df27['Tramos_Castigos'] = df27.apply(tramos_castigos, axis=1)
df27.head(10)
```

```python
#Creo una columna que me indique si el cliente tiene una mora hoy o no en Equifax o en R04, luego la aplico a todo el df27 retornando los resultado en la columna Mora_Hoy
def mora_hoy_saber(mora_hoy_o_no):
    if mora_hoy_o_no['Mora_EQFX_Hoy'] == "Si" or mora_hoy_o_no['Mora_R04_Hoy'] == "Si":
        return 'Si'
    else:
        return 'No'
    
df27['Mora_Hoy'] = df27.apply(mora_hoy_saber, axis=1)
df27.head(10)
```

```python
#Reviso si existen moras antiguas en los ultimos 12 meses en R04, si es asi retorno un Si, sino un No
def mora_antigua(numero_moras):
    if numero_moras['Deuda_DMora_12M'] > 0 or numero_moras['Deuda_DVencida_12M'] > 0 or numero_moras['Deuda_DCastigo_12M'] > 0:
        return 'Si'
    else:
        return 'No'
    
df27['Mora_Antigua_R04'] = df27.apply(mora_antigua, axis=1)
df27.head(10)
```

```python
#Reviso si existen moras antiguas en los ultimos 12 meses en Equifax, si es asi retorno un Si, sino un No. Luego aplico la funcion a todo el df27, generado la columna Mora_Antigua_EQFX donde se almacenan los resultados.
def mora_antigua_eqfx(numero_moras_eqfx):
    if numero_moras_eqfx['Mora_EQ_12_Meses'] > 0:
        return 'Si'
    else:
        return 'No'

df27['Mora_Antigua_EQFX'] = df27.apply(mora_antigua_eqfx, axis=1)
df27.head(10)
```

```python
#Reviso si existen moras antiguas en los ultimos 12 meses en R04 y Equifax, si es asi retorno un Si, sino un No. Luego aplico la funcion a todo el df27, generado la columna Mora_antigua_ambas donde se almacenan los resultados.
def mora_antigua_ambas(numero_moras_ambas):
    if numero_moras_ambas['Mora_Antigua_R04'] == 'Si' or numero_moras_ambas['Mora_Antigua_EQFX'] == 'Si':
        return 'Si'
    else:
        return 'No'
    
df27['Mora_antigua_ambas'] = df27.apply(mora_antigua_ambas, axis=1)
df27.head(10)
```

```python
#Reviso si el cliente tiene mora hoy o bien antigua, si es en ambas, en una si y en otra no, o en ninguna. En base a eso, genero una nueva columna llamada Tiene_Mora donde se almacenan los resultados.
def tiene_mora_o_no(tener_mora):
    if tener_mora['Mora_Hoy'] == 'Si' and tener_mora['Mora_antigua_ambas'] == 'Si':
        return 'Ambas'
    elif tener_mora['Mora_Hoy'] == 'Si' and tener_mora['Mora_antigua_ambas'] == 'No':
        return 'Mora Nueva'
    elif tener_mora['Mora_Hoy'] == 'No' and tener_mora['Mora_antigua_ambas'] == 'Si':
        return 'Mora Antigua'
    else:
        return 'Sin Mora'

df27['Tiene_Mora'] = df27.apply(tiene_mora_o_no, axis=1)
df27.head(10)
```

```python
#Reviso la maxima mora que hay entre las columnas de MTO_TOT_BIC y MTO_TOT_MCC_EQFX, y retorno el valor maximo entre las 2. Luego aplico la funcion a todo el df27, generado la columna Mora_Max donde se almacenan los resultados.
def mora_maxima(mora_max_):
    return mora_max_[['MTO_TOT_BIC','MTO_TOT_MCC_EQFX']].max(axis=0)

df27['Mora_Max'] = df27.apply(mora_maxima, axis=1)
df27.head(10)
```

```python
#tramifico la mora maxima, para poder hacer un analisis mas facil de la misma. Luego aplico la funcion a todo el df27, generado la columna Tramos_Mora_Max donde se almacenan los resultados.
def tramos_mora_max(tramos_moras_max):
    if tramos_moras_max['Mora_Max'] > 0 and tramos_moras_max['Mora_Max'] <= 10000:
        return '<= 10m'
    elif tramos_moras_max['Mora_Max'] > 10000 and tramos_moras_max['Mora_Max'] <= 30000:
        return 'Entre 10-30m'
    elif tramos_moras_max['Mora_Max'] > 30000 and tramos_moras_max['Mora_Max'] <= 50000:
        return 'Entre 30-50m'
    elif tramos_moras_max['Mora_Max'] > 50000 and tramos_moras_max['Mora_Max'] <= 100000:
        return 'Entre 50-100m'
    elif tramos_moras_max['Mora_Max'] > 100000:
        return '> 100m'
    else:
        return 'Sin Mora'
    
df27['Mora_Max_Tramos'] = df27.apply(tramos_mora_max, axis=1)
df27.head(10)
```
